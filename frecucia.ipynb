```python
def contar_frecuencia(lista, numeros):
    frecuencias = {num: [] for num in numeros}  # Diccionario para almacenar las frecuencias de cada número

    ultimo_indice = -1  # Último índice encontrado para cada número
    for indice, elemento in enumerate(lista):
        if elemento in numeros:
            if ultimo_indice == -1:
                distancia = indice + 1
            else:
                distancia = indice - ultimo_indice
            frecuencias[elemento].append(distancia)
            ultimo_indice = indice

    return frecuencias

# Ejemplo de uso:
numeros = [0, 1, 2, 3]
lista = [1, 2, 1, 3, 0, 3, 2, 0, 1, 2, 3, 0]
frecuencias = contar_frecuencia(lista, numeros)

for num in numeros:
    print(f"Número {num}: {frecuencias[num]}")
```
